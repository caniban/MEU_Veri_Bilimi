### Soru 1

Bir hedef noktasÄ±na yapÄ±lan atÄ±ÅŸlarÄ±n koordinatlarÄ± 5 kez Ã¶lÃ§Ã¼lmÃ¼ÅŸtÃ¼r.  
Hedef noktasÄ±nÄ±n bilinen koordinatÄ±:

$
L_{\text{gerÃ§ek}} = 100.0
$

AÅŸaÄŸÄ±daki dizide verilen 5 adet Ã¶lÃ§Ã¼len deÄŸeri kullanarak,  
$( L_{\text{gerÃ§ek}} )$ deÄŸerine gÃ¶re **KÃ¶k Karesel Ortalama Hata (KKOH)** hesaplayÄ±nÄ±z.

**FormÃ¼l:**


KKOH = $\sqrt{\frac{\sum_{i=1}^{n} (L_i - L_{\text{gerÃ§ek}})^2}{n}}$



**Ã–lÃ§Ã¼len deÄŸerler**
[99.8, 100.5, 100.2, 99.9, 100.1]




```python
import numpy as np

L_gercek = 100.0
# Verilen Ã¶lÃ§Ã¼len deÄŸerler
olcumler = np.array([99.8, 100.5, 100.2, 99.9, 100.1])
n = len(olcumler)

# 1. HatalarÄ±n HesaplanmasÄ± (v = L_i - L_gercek)
hatalar_v = olcumler - L_gercek

# 2. HatalarÄ±n Karelerinin HesaplanmasÄ± (v^2)
# ** operatÃ¶rÃ¼ kullanÄ±lÄ±r.
hatalar_kare_v2 = hatalar_v ** 2

# 3. HatalarÄ±n Karelerinin ToplamÄ±
hatalar_kare_toplami = np.sum(hatalar_kare_v2)

# 4. KKOH HesaplanmasÄ±

kkoh = np.sqrt(hatalar_kare_toplami / n)

# SonuÃ§larÄ±n YazdÄ±rÄ±lmasÄ±
print(f"Ã–lÃ§Ã¼m Verileri: {olcumler}")
print(f"GerÃ§ek DeÄŸer (L_gercek): {L_gercek}")
print(f"Ã–lÃ§Ã¼ SayÄ±sÄ± (n): {n}")
print("--------------------------------------------------")
print(f"Hatalar (v = L_i - L_gercek): {hatalar_v}")
print(f"HatalarÄ±n Kareleri (v^2): {hatalar_kare_v2}")
print(f"HatalarÄ±n Kareleri ToplamÄ± (Sum[v^2]): {hatalar_kare_toplami:.2f}")
print("--------------------------------------------------")
print(f"KÃ¶k Karesel Ortalama Hata (KKOH): {kkoh:.2f}")
```

### Soru 2

Bir araÃ§, sabit hÄ±zla  V = 60 $\text{km/saat}$ hÄ±zla hareket etmektedir.  

AÅŸaÄŸÄ±daki dizide verilen farklÄ± zaman aralÄ±klarÄ±nda (saat cinsinden) aracÄ±n baÅŸlangÄ±Ã§ noktasÄ±na gÃ¶re aldÄ±ÄŸÄ± yolu (konumu) hesaplayÄ±nÄ±z.

**FormÃ¼l:**

$
X = V \cdot t
$



**Zaman aralÄ±klarÄ± (saat cinsinden)**
[0.5, 1, 1.5, 2, 2.5]




```python
# Soru 2'nin Ã§Ã¶zÃ¼mÃ¼
import numpy as np

V = 60.0  # km/saat
zamanlar_t = np.array([0.5, 1.0, 1.5, 2.0, 2.5]) # saat

# KonumlarÄ±n hesaplanmasÄ± (X = V * t)
# NumPy, skaler (V) ile diziyi (zamanlar_t) otomatik olarak her eleman iÃ§in Ã§arpar.
konumlar_x = V * zamanlar_t

# SonuÃ§larÄ±n YazdÄ±rÄ±lmasÄ± 
print(f"Sabit HÄ±z (V): {V} km/saat")
print(f"Zaman AralÄ±klarÄ± (t): {zamanlar_t} saat")
print("--------------------------------------------------")
print(f"Hesaplanan Konumlar (X): {konumlar_x} km")
```

### Soru 3

DÃ¼zlemde sabit bir nokta $ P_1 = (5, 10) $ verilmiÅŸtir.  
AyrÄ±ca aÅŸaÄŸÄ±daki dizilerde koordinatlarÄ± verilen 4 adet nokta $ P_2 = (x_2, y_2) $ bulunmaktadÄ±r.  

$ P_1 $ ve her bir $ P_2 $ noktasÄ± arasÄ±ndaki **Ã–klid uzaklÄ±ÄŸÄ±nÄ±** hesaplayÄ±nÄ±z.

**Ã–klid UzaklÄ±k FormÃ¼lÃ¼:**

$
d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
$


#### DiÄŸer noktalar (P2)
x = [2, 8, 6, 5]

y= [3, 10, 14, 7]


```python
#Soru 3'Ã¼n Ã§Ã¶zÃ¼mÃ¼
import numpy as np

# Sabit P1 noktasÄ±nÄ±n koordinatlarÄ± (VarsayÄ±lan olarak Ã¶nceki sorulardan alÄ±ndÄ±)
P1_x = 5.0
P1_y = 10.0

# P2 noktalarÄ±nÄ±n koordinatlarÄ±
P2_x_koordinatlari = np.array([2, 8, 6, 5])
P2_y_koordinatlari = np.array([3, 10, 14, 7])

# 1. x ve y farklarÄ±nÄ±n hesaplanmasÄ± (dx = x2 - x1, dy = y2 - y1)
dx = P2_x_koordinatlari - P1_x
dy = P2_y_koordinatlari - P1_y

# 2. FarklarÄ±n karelerinin hesaplanmasÄ± (dx^2 ve dy^2)
dx_kare = dx ** 2
dy_kare = dy ** 2

# 3. Karelerin toplamÄ±nÄ±n hesaplanmasÄ± (dx^2 + dy^2)
kareler_toplami = dx_kare + dy_kare

# 4. KarekÃ¶kÃ¼n alÄ±nmasÄ± (Ã–klid UzaklÄ±ÄŸÄ±)
uzakliklar = np.sqrt(kareler_toplami)

# SonuÃ§larÄ±n YazdÄ±rÄ±lmasÄ±
print(f"Sabit Nokta P1: ({P1_x}, {P1_y})")
print(f"P2 x KoordinatlarÄ±: {P2_x_koordinatlari}")
print(f"P2 y KoordinatlarÄ±: {P2_y_koordinatlari}")
print("--------------------------------------------------")
print(f"x FarklarÄ± (dx): {dx}")
print(f"y FarklarÄ± (dy): {dy}")
print(f"dx^2: {dx_kare}")
print(f"dy^2: {dy_kare}")
print(f"dx^2 + dy^2: {kareler_toplami}")
print("--------------------------------------------------")
print(f"Hesaplanan Ã–klid UzaklÄ±klarÄ± (d): {uzakliklar}")
```
